```python
# Render our plots inline
%matplotlib inline

import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
birthrate = pd.read_csv("all_imp.csv")
```

```python
birthrate.isna().sum()
```

```python
birthrate['any'].fillna(0, inplace=True)
birthrate['mmore'].fillna(0, inplace=True)
birthrate['class'].fillna(0, inplace=True)
birthrate['edu_level'].fillna(0, inplace=True)
birthrate['newspaper'].fillna(0, inplace=True)
birthrate['born'].fillna(1, inplace=True)
birthrate['boys_died'].fillna(0, inplace=True)
birthrate['girls_died'].fillna(0, inplace=True)

birthrate.dropna(axis=0, subset=['method'], inplace=True)
birthrate.dropna(axis=0, subset=['with_partner'], inplace=True)
birthrate.dropna(axis=0, subset=['first_child_gender'], inplace=True)
birthrate.dropna(axis=0, subset=['children_away'], inplace=True)
```

```python
birthrate['any'].replace(2,0, inplace=True)
birthrate['mmore'].replace(2,0, inplace=True)
birthrate['any'].replace(8,0, inplace=True)
birthrate['mmore'].replace(8,0, inplace=True)

birthrate['more'] = birthrate['any'] + birthrate['mmore']
```

```python
birthrate['school'].replace(2,0, inplace=True)
birthrate['working'].replace(2,0, inplace=True)
birthrate['newspaper'].replace(2,0, inplace=True)
birthrate['tv'].replace(2,0, inplace=True)
birthrate['asha'].replace(2,0, inplace=True)
birthrate['grameen'].replace(2,0, inplace=True)
birthrate['mclub'].replace(2,0, inplace=True)
birthrate['more'].replace(2,0, inplace=True)
birthrate['method'].replace(2,0, inplace=True)
birthrate['with_partner'].replace(2,0, inplace=True)
birthrate['children_away'].replace(2,0, inplace=True)
```

```python
birthrate.dropna(axis=0, subset=['age'], inplace=True)
```

```python
birthrate.isna().sum()
```

```python
len(birthrate)
for col in birthrate.columns:
    print(birthrate[col].value_counts())
```

```python
birthrate['division'] = birthrate['division'].astype('category')
birthrate['domain'] = birthrate['domain'].astype('category')
birthrate['first_child_gender'] = birthrate['first_child_gender'].astype('category')
birthrate['born'] = birthrate['born'].astype('category')
```

```python
birthrate = pd.get_dummies(birthrate)
```

```python
birthrate.to_csv("allimp_modified.csv", index=False)
```
