```python
# Render our plots inline
%matplotlib inline

import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
birthrate = pd.read_csv("all_imp.csv")
```

```python
birthrate.isna().sum()
```

```python
birthrate['any'].fillna(0, inplace=True)
birthrate['mmore'].fillna(0, inplace=True)
birthrate['class'].fillna(0, inplace=True)
birthrate['born_place'].fillna(1, inplace=True)
birthrate['boys_died'].fillna(0, inplace=True)
birthrate['girls_died'].fillna(0, inplace=True)

birthrate.dropna(axis=0, subset=['method'], inplace=True)
birthrate.dropna(axis=0, subset=['first_child_gender'], inplace=True)
birthrate.dropna(axis=0, subset=['children_away'], inplace=True)
```

```python
birthrate['any'].replace(2,0, inplace=True)
birthrate['mmore'].replace(2,0, inplace=True)
birthrate['any'].replace(8,0, inplace=True)
birthrate['mmore'].replace(8,0, inplace=True)

birthrate['want_child'] = birthrate['any'] + birthrate['mmore']
```

```python
birthrate.dropna(axis=0, subset=['age'], inplace=True)
```
