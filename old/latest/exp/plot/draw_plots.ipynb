```python
# Render our plots inline
%matplotlib inline

import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
birthrate = pd.read_csv("birth_final_cat.csv")
```

```python
sample = birthrate.sample(frac=.01, axis=0)
```

```python
dec_name='birth'
feat_names = list(birthrate.columns)
feat_names.remove(dec_name)
```

```python
for name in feat_names:
    sample.plot(x=name, y=dec_name, style=['o','rx'])
```
